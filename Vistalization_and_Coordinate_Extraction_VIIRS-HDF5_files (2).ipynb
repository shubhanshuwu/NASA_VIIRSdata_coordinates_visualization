```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Define the coordinates of the four corners for each set
sets = [
    {
        "top_left": (30.0, 70.0),
        "top_right": (30.0, 79.99583333332424),
        "bottom_left": (20.004166666667235, 70.0),
        "bottom_right": (20.004166666667235, 79.99583333332424)
    },
    {
        "top_left": (20.0, 70.0),
        "top_right": (20.0, 79.99583333332424),
        "bottom_left": (10.004166666667235, 70.0),
        "bottom_right": (10.004166666667235, 79.99583333332424)
    },
    {
        "top_left": (30.0, 80.0),
        "top_right": (30.0, 89.99583333332424),
        "bottom_left": (20.004166666667235, 80.0),
        "bottom_right": (20.004166666667235, 89.99583333332424)
    },
    {
        "top_left": (20.0, 80.0),
        "top_right": (20.0, 89.99583333332424),
        "bottom_left": (10.004166666667235, 80.0),
        "bottom_right": (10.004166666667235, 89.99583333332424)
    }
]

# Create a new figure
plt.figure(figsize=(10, 8))

# Create a GeoAxes with the PlateCarree projection
ax = plt.axes(projection=ccrs.PlateCarree())

# Add coastlines
ax.coastlines()

# Iterate over each set
for idx, set_coords in enumerate(sets, start=1):
    # Plot the box for each set
    ax.plot([set_coords['top_left'][1], set_coords['top_right'][1],
             set_coords['bottom_right'][1], set_coords['bottom_left'][1],
             set_coords['top_left'][1]],
            [set_coords['top_left'][0], set_coords['top_right'][0],
             set_coords['bottom_right'][0], set_coords['bottom_left'][0],
             set_coords['top_left'][0]],
            linestyle='-', color='red', linewidth=1, transform=ccrs.PlateCarree())
    
    # Add labels to the center of each box
    center_lat = (set_coords['top_left'][0] + set_coords['bottom_right'][0]) / 2
    center_lon = (set_coords['top_left'][1] + set_coords['bottom_right'][1]) / 2
    ax.text(center_lon, center_lat, f'Set {idx}', transform=ccrs.PlateCarree(), fontsize=10, color='black')

# Set extent to zoom out
ax.set_extent([60, 100, 0, 40], crs=ccrs.PlateCarree())

# Set title
plt.title('Geographical Coordinates of Grid Corners')

# Show the map
plt.show()
```
